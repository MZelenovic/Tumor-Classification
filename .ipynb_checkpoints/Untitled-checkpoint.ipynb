```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split,GridSearchCV,StratifiedKFold
from sklearn import tree
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, StackingClassifier
from sklearn.svm import SVC
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score
import seaborn as sns
from sklearn.linear_model import LogisticRegression
```

```python
tumor_data = pd.read_csv('Cancer_Data.csv')
tumor_data.head()
```

```python
tumor_data['diagnosis'] = tumor_data['diagnosis'].map({'M': 1, 'B': 0})
tumor_data.head()
```

```python
correlation = tumor_data.corr()
correlation
```

```python
plt.figure(figsize=(10, 8))
sns.heatmap(correlation, annot=False, cmap ='PuOr')
plt.title('Correlation Heatmap', fontsize=20)
plt.show()
```

```python
selected_features = ['radius_mean', 'texture_mean', 'perimeter_mean', 'area_mean', 'smoothness_mean', 'compactness_mean', 'concavity_mean', 'concave points_mean', 'symmetry_mean', 'fractal_dimension_mean' ]
pair_plot = sns.pairplot(tumor_data[selected_features], diag_kind='kde')
plt.suptitle('Pair Plot of Selected Features', y= 1.02, fontsize=24)
plt.show()
```

```python
# Machine Learnig Classification
y = tumor_data['diagnosis']
X = tumor_data.drop(['id', 'diagnosis', 'Unnamed: 32'], axis = 1)
print(y.value_counts())
# Normalising predictors

scaler = StandardScaler()
scaler.fit(X)
X = scaler.transform(X)

X_train,X_test,y_train,y_test = train_test_split(X,y,train_size = 0.8, random_state=11)
```

```python
# Random Forest Classifier
rf = RandomForestClassifier() #For now we will use default hyperparams
rf.fit(X_train,y_train)

y_pred_test = rf.predict(X_test)

print('RF initial Test Accuracy: ',rf.score(X_test,y_test))
print('RF initial Test Recall: ', recall_score(y_test,y_pred_test))
```

```python
#Parameter tuning using gridseachCV

depths = list(range(2,11))
params = {'criterion' : ['gini','entropy'],'max_depth':depths,'random_state':[12]}

rf_cv = GridSearchCV(rf,params,scoring='recall')
rf_cv.fit(X_train,y_train)
#Note we are using recall as it's the ratio of correct positive predictions classifications made by the model to all 
#actual positives
#For a cancer patient a false positive is much less harmful than a false negative, so we will be using recall as a higher
#recall score will mean less false negatives
#note that we will also consider accuracy as well, but recall will be our priority


print(rf_cv.best_estimator_)
```

```python
rf_best = rf_cv.best_estimator_
rf_best.fit(X_train,y_train)

rf_best_pred = rf_best.predict(X_test)

print('Best RF Test Accuracy: ',rf_best.score(X_test,y_test))
print('Best RF Test Recall: ', recall_score(y_test,rf_best_pred))
```

```python
# Support Vector Classifier
svc = SVC()
svc.fit(X_train,y_train)

svc_pred = svc.predict(X_test)

print('SVC initial Test Accuracy: ',svc.score(X_test,y_test))
print('SVC initial Test Recall: ', recall_score(y_test,svc_pred))
```

```python
C_list = list(np.linspace(0.1,2,num=20))

kernels = ['rbf','linear','poly','sigmoid']

params = {'C':C_list,'kernel':kernels,'random_state':[0]}

svc_CV = GridSearchCV(svc,params,scoring='recall')
svc_CV.fit(X_train,y_train)

print(svc_CV.best_estimator_)
```

```python
svc_best = svc_CV.best_estimator_
svc_best.fit(X_train,y_train)

svc_best_pred = svc_best.predict(X_test)

print('Best RF Test Accuracy: ',svc_best.score(X_test,y_test))
print('Best RF Test Recall: ', recall_score(y_test,svc_best_pred))
```

```python
# Logistic Regression 
lr = LogisticRegression(solver='liblinear',random_state=12)
lr.fit(X_train,y_train)
y_pred_test = lr.predict(X_test)

print('LR initial Test Accuracy: ',lr.score(X_test,y_test))
print('LR initial Test Recall: ', recall_score(y_test,y_pred_test))
```

```python
C_list = list(np.linspace(0.01,10,num=20))  # Različite vrednosti za hiperparametar C

params = {'C': C_list, 'penalty': ['l1', 'l2'], 'random_state': [12]}

logreg_CV = GridSearchCV(lr, params, scoring='recall')
logreg_CV.fit(X_train,y_train)

print(logreg_CV.best_estimator_)
```
